```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import util
```

```python
img = cv2.imread("./img/jpeg/im1.jpeg")
```

```python
# downscale image
times = 8
img = cv2.resize(img, (img.shape[1] // times, img.shape[0] // times))
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
b, g, r = cv2.split(img)
# show the red and green channels
plt.figure()
plt.imshow(r)
plt.figure()
plt.imshow(g)
```

```python
# subtract the green channel from the red channel
r_minus_g = util.overflow_subtract(r, g)
plt.figure()
plt.imshow(r_minus_g, cmap="gray")
g_minus_r = util.overflow_subtract(g, r)
plt.figure()
plt.imshow(g_minus_r, cmap="gray")
```

```python
green_centers = util.blob_detect_center(r_minus_g)
red_centers = util.blob_detect_center(g_minus_r)
```

```python
# draw points for each center
for center in red_centers:
    cv2.circle(img_rgb, center, 5, (255, 0, 0), -1)

for center in green_centers:
    cv2.circle(img_rgb, center, 5, (0, 255, 0), -1)
    
plt.imshow(img_rgb)
```
